# Лабораторная работа №7. "Модели конкуренции"
## Вариант 27

Межвидовая конкуренция. В этом случае исследуется конкуренция популяций, потребляющих общий ресурс. Пусть $N_1$ и $N_2$ - численности конкурирующих популяций. Модель (Лотки-Вольтерры) выражается уравнениями:
$$
\begin{cases}
    \frac{dN_1}{dt} = r_1N_1\frac{K_1-N_1-\alpha_{12}N_2}{K_1} \\
    \frac{dN_2}{dt} = r_2N_2\frac{K_2-N_2-\alpha_{21}N_1}{K_2}
\end{cases}
$$

Параметры $\alpha_{12}$ и $\alpha_{21}$ отражают интенсивность межвидовой конкуренции. <br> 
Главный вопрос, который интересует исследователя межвидовой конкуренции, - при каких условиях увеличивается или уменьшается численность каждого вида? <br>
Данная модель предсказывает следующие режимы эволюции взаимодействующих популяций: устойчивое сосуществование или полное вытеснение одной из них. <br>
Построить в фазовой плоскости $(N_1, N_2)$ границы зон, разделяющих какие-либо два режима эволюции конкурирующих популяций. Остальные параметры модели выбрать произвольно. Учесть при этом, что режим устойчивого сосуществования популяций может в принципе реализоваться только при $\alpha_{12}<1$, $\alpha_{21}<1$. <br>
Визуализировать в виде графиков зависимости от $t$ и фазовых диаграмм, на которых указать особые точки, изоклины и направления на траекториях. 

## Решение 
Все коэффициенты неотрицательные <br><br>


#### Найдём особые точки:
$$
\begin{cases}
    r_1N_1 \frac{K_1-N_1-\alpha_{12}N_2}{K_1} = 0\\
    r_2N_2 \frac{K_2-N_2-\alpha_{21}N_1}{K_2} = 0
\end{cases}

\implies 

\begin{cases}
    N_1(r_1K_1-r_1N_1-\alpha_{12}r_1N_2) = 0 \\ 
    N_2(r_2K_2-r_2N_2-\alpha_{21}r_2N_1) = 0
\end{cases}
$$

Особая точка №1
$$
\begin{align*}
    N_1^{(1)} = 0 \\
    N_2^{(1)} = 0
\end{align*}
$$

Особая точка №2
$$
\begin{align*}
    N_1^{(2)} = 0 \\
    r_2K_2-r_2N_2 =0
\end{align*}

\implies 

\begin{align*}
    N_1^{(2)} &= 0 \\
    N_2^{(2)} &= K_2
\end{align*}
$$

Особая точка №3

$$
\begin{align*}
    r_1K_1-r_1N_1 = 0 \\
    N_2^{(3)} = 0
\end{align*}

\implies 

\begin{align*}
    N_1^{(3)} &= K_1 \\
    N_2^{(3)} &= 0
\end{align*}
$$

Особая точка №4
$$
\begin{cases}
    K_1-N_1-\alpha_{12}N_2 = 0 \\ 
    K_2-N_2-\alpha_{21}N_1 = 0
\end{cases}

\implies

\begin{cases}
    N_1 =  K_1 - \alpha_{12}N_2 \\ 
    K_2-N_2-\alpha_{21}(K_1 - \alpha_{12}N_2 ) = 0
\end{cases}

\implies

\begin{cases}
    N_1 =  K_1 - \alpha_{12}N_2 \\ 
    K_2-N_2-\alpha_{21}K_1 + \alpha_{21}\alpha_{12}N_2 = 0
\end{cases}

\implies
$$

$$
\begin{cases}
    N_1 =  K_1 - \alpha_{12}N_2 \\ 
    N_2 = \frac{\alpha_{21}K_1 - K_2}{\alpha_{21}\alpha_{12} - 1}
\end{cases}

\implies

\begin{cases}
    N_1^{(4)} =  K_1 - \frac{\alpha_{12}\alpha_{21}K_1 - \alpha_{12}K_2}{\alpha_{21}\alpha_{12} - 1} =  \frac{\alpha_{12}K_2 -K_1}{\alpha_{21}\alpha_{12} - 1} \\ 
    N_2^{(4)} = \frac{\alpha_{21}K_1 - K_2}{\alpha_{21}\alpha_{12} - 1}
\end{cases}
$$

#### Определим фазовый портрет 
$$J = 
\begin{pmatrix}\frac{r_1(K_1 - 2N_1-N_2\alpha_{12})}{K_1} & -\frac{r_1\alpha_{12}N_1}{K_1} \\ -\frac{r_2\alpha_{21}N_2}{K_2} & \frac{r_2(K_2 - 2N_2-N_1\alpha_{21})}{K_2}\end{pmatrix}
$$



Особая точка $(0,0)$
$$
\begin{bmatrix}
    r_1  - \lambda & 0 \\
    0 & r_2  - \lambda 
\end{bmatrix}
=
(r_1  - \lambda)(r_2 - \lambda ) = 0
\implies
\begin{cases}
    \lambda_{1} = r_1 \\
    \lambda_{2} = r_2
\end{cases}
\implies
неустойчивый \quad узел
$$

Особая точка $(0, K_2)$
$$
\begin{bmatrix}
    \frac{r_1K_1  -r_1K_2\alpha_{12}}{K_1}-\lambda & 0 \\
    -r_2\alpha_{21} & \frac{r_2K_2 -2r_2K_2}{K_2} - \lambda 
\end{bmatrix}
= (\frac{r_1K_1  -r_1K_2\alpha_{12}}{K_1}-\lambda)(\frac{r_2K_2 -2r_2K_2}{K_2} - \lambda) = 0 
\implies
\begin{cases}
    \lambda_{1} = r_1 - \frac{r_1K_2\alpha_{12}}{K_1} \\
    \lambda_{2} = -r_2
\end{cases}

\implies
\begin{cases}
    устойчивый \quad узел, \quad K_1 < K_2\alpha_{12} \\
    седло, \quad иначе
\end{cases}
$$

Особая точка $(K_1, 0)$
$$
\begin{bmatrix}
    \frac{r_1K_1 -2r_1K_1}{K_1} - \lambda  & -r_1\alpha_{12}\\
    0 & \frac{r_2K_2  -r_2K_1\alpha_{21}}{K_2}-\lambda 
\end{bmatrix}
= (\frac{r_1K_1 -2r_1K_1}{K_1} - \lambda)(\frac{r_2K_2  -r_2K_1\alpha_{21}}{K_2}-\lambda) = 0 
\implies
\begin{cases}
    \lambda_{1} = -r_1  \\
    \lambda_{2} = r_2  -\frac{r_2K_1\alpha_{21}}{K_2}
\end{cases}

\implies
\begin{cases}
    устойчивый \quad узел, \quad K_2 < K_1\alpha_{21} \\
    седло, \quad иначе
\end{cases}
$$

Особая точка $( \frac{\alpha_{12}K_2 -K_1}{\alpha_{21}\alpha_{12} - 1}, \frac{\alpha_{21}K_1 - K_2}{\alpha_{21}\alpha_{12} - 1})$


$
\begin{bmatrix}
   -\frac{r_1}{K_1}N_1 - \lambda  & -\frac{r_1\alpha_{12}}{K_1}N_1 \\
   -\frac{r_2\alpha_{21}}{K_2}N_2 & -\frac{r_2}{K_2}N_2-\lambda 
\end{bmatrix}= 
 (-\frac{r_1}{K_1}N_1 - \lambda)(-\frac{r_2}{K_2}N_2-\lambda ) -\frac{r_1r_2\alpha_{12}\alpha_{21}}{K_1K_2}N_1N_2 = \lambda^2 + \lambda(\frac{r_1}{K_1}N_1+\frac{r_2}{K_2}N_2) + \frac{r_1r_2}{K_1K_2}N_1N_2 -\frac{r_1r_2\alpha_{12}\alpha_{21}}{K_1K_2}N_1N_2 = 0
$

$$
\lambda^2 + \lambda\frac{r_1N_1K_2+r_2N_2K_1}{K_1K_2} + \frac{r_1r_2N_1N_2(1-\alpha_{12}\alpha_{21})}{K_1K_2}  = 0
$$

$$
\lambda_{1}\lambda_{2} = \frac{r_1r_2N_1N_2(1-\alpha_{12}\alpha_{21})}{K_1K_2}
$$
$$\frac{r_1r_2N_1N_2(1-\alpha_{12}\alpha_{21})}{K_1K_2} > 0$$
$$r_1r_2N_1N_2(1-\alpha_{12}\alpha_{21}) > 0$$
$$\frac{(\alpha_{12}K_2 -K_1)(\alpha_{21}K_1 - K_2)(1-\alpha_{12}\alpha_{21})}{(\alpha_{21}\alpha_{12} - 1)^2} > 0$$
$$(\alpha_{12}K_2 -K_1)(\alpha_{21}K_1 - K_2) > 0$$
$$\alpha_{21}\alpha_{12}K_1K_2-\alpha_{12}K_2^2-\alpha_{21}K_1^2+K_1K_2 >0$$
$$K_1K_2 > \alpha_{21}K_1^2 + \alpha_{12}K_2^2 - \alpha_{21}\alpha_{12}K_1K_2$$
$$K_1 > \frac{\alpha_{21}K_1^2 + \alpha_{12}K_2^2 - \alpha_{21}\alpha_{12}K_1K_2}{K_2}$$
От сюда следует
$$
\begin{cases}
    узел \quad при \quad K_1 > \frac{\alpha_{21}K_1^2 + \alpha_{12}K_2^2 - \alpha_{21}\alpha_{12}K_1K_2}{K_2} \\
    седло \quad иначе
\end{cases}
$$

$$
\lambda_{1} + \lambda_{2} = - \frac{r_1N_1K_2+r_2N_2K_1}{K_1K_2}
$$

$$
- \frac{r_1N_1K_2+r_2N_2K_1}{K_1K_2} < 0
$$

$$ \frac{\alpha_{21}K_1^2 + \alpha_{12}K_2^2 - \alpha_{21}\alpha_{12}K_1K_2}{K_2} < K_1 \quad - \quad устойчивый \quad узел$$

#### Определим изоклины
$$\begin{cases}
    r_1N_1 \frac{K_1-N_1-\alpha_{12}N_2}{K_1} = 0\\
    r_2N_2 \frac{K_2-N_2-\alpha_{21}N_1}{K_2} = 0
\end{cases}$$
Уравнение изоклин горизонтальных касательных (dN₁/dt = 0):
$$ N_1 = 0 \quad и \quad N_2 =\frac{K_1-N_1}{\alpha_{12}} $$
Уравнение изоклин вертикальных касательных (dN₂/dt = 0):
$$N_2 = 0 \quad и \quad  N_2= K_2-\alpha_{21}N_1$$

Изоклины разделяют фазовую плоскость на зоны с различными направлениями изменения численности популяций. Точки пересечения изоклин являются особыми точками системы.

#### Границы зон в фазовой плоскости

Границы зон определяются изоклинами, которые разделяют области с разными режимами эволюции:
- **Зона роста N₁ и N₂**: обе популяции увеличиваются
- **Зона роста N₁, убывания N₂**: вид 1 вытесняет вид 2
- **Зона убывания N₁, роста N₂**: вид 2 вытесняет вид 1  
- **Зона убывания N₁ и N₂**: обе популяции уменьшаются

Границы определяются условиями:
- $dN_1/dt = 0$: $N_1 = 0$ или $N_2 = \frac{K_1 - N_1}{\alpha_{12}}$
- $dN_2/dt = 0$: $N_2 = 0$ или $N_2 = K_2 - \alpha_{21}N_1$

Условие устойчивого сосуществования: $\alpha_{12} < 1$ и $\alpha_{21} < 1$, при этом точка пересечения изоклин $(N_1^*, N_2^*)$ должна быть устойчивой особой точкой.


```python
# Библиотеки 
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Модель Лотки-Вольтерры
def competition_model(N, t, r1, r2, K1, K2, alpha12, alpha21):
    N1, N2 = N
    dN1_dt = r1 * N1 * (K1 - N1 - alpha12 * N2) / K1
    dN2_dt = r2 * N2 * (K2 - N2 - alpha21 * N1) / K2
    return [dN1_dt, dN2_dt] 

# Фазовый портрет с границами зон
def plot_phase_portrait(ax, r1, r2, K1, K2, alpha12, alpha21, title):
    # Определение особых точек
    points = []
    point_labels = []
    
    # Точка (0, 0)
    points.append((0, 0))
    point_labels.append('(0,0) - Неустойчивый узел')
    
    # Точка (K1, 0)
    points.append((K1, 0))
    if K2 < alpha21 * K1:
        point_labels.append(f'({K1:.2f},0) - Устойчивый узел')
    else:
        point_labels.append(f'({K1:.2f},0) - Седло')
    
    # Точка (0, K2)
    points.append((0, K2))
    if K1 < alpha12 * K2:
        point_labels.append(f'(0,{K2:.2f}) - Устойчивый узел')
    else:
        point_labels.append(f'(0,{K2:.2f}) - Седло')
    
    # Точка пересечения изоклин (если существует)
    denom = 1 - alpha12 * alpha21
    if abs(denom) > 1e-10:
        N1_star = (K1 - alpha12 * K2) / denom
        N2_star = (K2 - alpha21 * K1) / denom
        if N1_star > 0 and N2_star > 0:
            points.append((N1_star, N2_star))
            # Проверка устойчивости
            cond = K1 * K2 > alpha21 * K1**2 + alpha12 * K2**2 - alpha12 * alpha21 * K1 * K2
            if cond:
                point_labels.append(f'({N1_star:.2f},{N2_star:.2f}) - Устойчивый узел')
            else:
                point_labels.append(f'({N1_star:.2f},{N2_star:.2f}) - Седло')
    
    # Создание сетки для фазового портрета
    N1_max = max(K1 * 1.2, max([p[0] for p in points if p[0] > 0]) * 1.2) if points else K1 * 1.2
    N2_max = max(K2 * 1.2, max([p[1] for p in points if p[1] > 0]) * 1.2) if points else K2 * 1.2
    
    N1 = np.linspace(0.01, N1_max, 25)
    N2 = np.linspace(0.01, N2_max, 25)
    N1_grid, N2_grid = np.meshgrid(N1, N2)
    
    # Вычисление скоростей
    dN1_dt = np.zeros_like(N1_grid)
    dN2_dt = np.zeros_like(N2_grid)
    
    for i in range(len(N1)):
        for j in range(len(N2)):
            dN = competition_model([N1_grid[i,j], N2_grid[i,j]], 0, 
                                 r1, r2, K1, K2, alpha12, alpha21)
            dN1_dt[i,j] = dN[0]
            dN2_dt[i,j] = dN[1]
    
    # Нормализация стрелок
    magnitude = np.sqrt(dN1_dt**2 + dN2_dt**2)
    magnitude = np.where(magnitude < 1e-10, 1, magnitude)  # Избегаем деления на ноль
    dN1_dt_norm = dN1_dt / magnitude
    dN2_dt_norm = dN2_dt / magnitude
    
    # Построение векторного поля
    ax.quiver(N1_grid, N2_grid, dN1_dt_norm, dN2_dt_norm, 
              color='blue', alpha=0.5, scale=25, width=0.003)
    
    # Изоклины (границы зон)
    N1_iso = np.linspace(0, N1_max, 200)
    
    # Изоклина dN1/dt = 0: N2 = (K1 - N1)/alpha12 (для N1 < K1)
    mask1 = N1_iso < K1
    N2_iso1 = np.where(mask1, (K1 - N1_iso) / alpha12, np.nan)
    ax.plot(N1_iso[mask1], N2_iso1[mask1], 'r-', linewidth=2, label='Изоклина dN₁/dt=0')
    
    # Изоклина dN2/dt = 0: N2 = K2 - alpha21*N1 (для N1 < K2/alpha21)
    mask2 = N1_iso < K2 / alpha21 if alpha21 > 0 else np.ones_like(N1_iso, dtype=bool)
    N2_iso2 = np.where(mask2, K2 - alpha21 * N1_iso, np.nan)
    ax.plot(N1_iso[mask2], N2_iso2[mask2], 'g-', linewidth=2, label='Изоклина dN₂/dt=0')
    
    # Оси координат (также изоклины)
    ax.axvline(x=0, color='black', linewidth=1.5, linestyle='--', alpha=0.7)
    ax.axhline(y=0, color='black', linewidth=1.5, linestyle='--', alpha=0.7)
    
    # Особые точки
    colors_pt = ['red', 'green', 'blue', 'purple', 'orange']
    markers_pt = ['o', 's', '^', 'D', 'v']
    for i, (x, y) in enumerate(points):
        if x >= 0 and y >= 0:
            ax.plot(x, y, marker=markers_pt[i % len(markers_pt)], 
                   markersize=10, color=colors_pt[i % len(colors_pt)], 
                   markeredgecolor='black', markeredgewidth=1.5,
                   label=point_labels[i], zorder=5)
    
    # Траектории из разных начальных условий
    t = np.linspace(0, 30, 2000)
    initial_conditions = [
        [0.2*K1, 0.2*K2], [0.5*K1, 0.1*K2], [0.1*K1, 0.5*K2],
        [0.7*K1, 0.3*K2], [0.3*K1, 0.7*K2], [0.6*K1, 0.6*K2]
    ]
    
    for N0 in initial_conditions:
        try:
            solution = odeint(competition_model, N0, t, 
                           args=(r1, r2, K1, K2, alpha12, alpha21))
            # Фильтруем только положительные значения
            valid = (solution[:, 0] > 0) & (solution[:, 1] > 0) & \
                   (solution[:, 0] < N1_max) & (solution[:, 1] < N2_max)
            if np.any(valid):
                ax.plot(solution[valid, 0], solution[valid, 1], 
                       'darkred', alpha=0.6, linewidth=1.5)
                ax.plot(solution[0, 0], solution[0, 1], 
                       'ko', markersize=4, zorder=6)
        except:
            pass
    
    ax.set_xlabel('N₁ (Популяция вида 1)', fontsize=11)
    ax.set_ylabel('N₂ (Популяция вида 2)', fontsize=11)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.legend(loc='best', fontsize=8)
    ax.grid(True, alpha=0.3)
    ax.set_xlim(-0.05*N1_max, N1_max)
    ax.set_ylim(-0.05*N2_max, N2_max)

def plot_time_series(ax, r1, r2, K1, K2, alpha12, alpha21, title):
    """Построение временных рядов"""
    t = np.linspace(0, 30, 2000)
    
    # Разные начальные условия
    initial_conditions = [
        [0.3*K1, 0.3*K2], 
        [0.7*K1, 0.2*K2],
        [0.2*K1, 0.7*K2]
    ]
    linestyles = ['-', '--', '-.']
    
    for i, N0 in enumerate(initial_conditions):
        try:
            solution = odeint(competition_model, N0, t, 
                           args=(r1, r2, K1, K2, alpha12, alpha21))
            ax.plot(t, solution[:, 0], color='red', 
                   linestyle=linestyles[i], linewidth=2,
                   label=f'N₁(t), N₀=({N0[0]:.2f},{N0[1]:.2f})')
            ax.plot(t, solution[:, 1], color='blue', 
                   linestyle=linestyles[i], linewidth=2,
                   label=f'N₂(t), N₀=({N0[0]:.2f},{N0[1]:.2f})')
        except:
            pass
    
    ax.set_xlabel('Время t', fontsize=11)
    ax.set_ylabel('Численность', fontsize=11)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.legend(loc='best', fontsize=8)
    ax.grid(True, alpha=0.3)

# Параметры для варианта 27
# Учитываем условие: α₁₂ < 1, α₂₁ < 1 для устойчивого сосуществования
params = {
    "Устойчивое сосуществование": {
        "r1": 1.2, "r2": 1.0, "K1": 1.0, "K2": 1.0,
        "alpha12": 0.6, "alpha21": 0.4  # Оба < 1
    },
    "Победа вида 1": {
        "r1": 1.0, "r2": 0.8, "K1": 1.0, "K2": 0.8,
        "alpha12": 0.5, "alpha21": 1.2  # α₂₁ > 1
    },
    "Победа вида 2": {
        "r1": 0.8, "r2": 1.0, "K1": 0.8, "K2": 1.0,
        "alpha12": 1.2, "alpha21": 0.5  # α₁₂ > 1
    },
    "Бистабильность": {
        "r1": 1.0, "r2": 1.0, "K1": 1.0, "K2": 1.0,
        "alpha12": 0.8, "alpha21": 0.8  # Оба < 1, но близки к 1
    }
}

# Построение графиков
fig, axes = plt.subplots(4, 2, figsize=(16, 20))
plt.rcParams['font.size'] = 10

for i, (regime, param) in enumerate(params.items()):
    # Фазовый портрет
    plot_phase_portrait(axes[i, 0], 
                       param["r1"], param["r2"], 
                       param["K1"], param["K2"],
                       param["alpha12"], param["alpha21"],
                       f"Фазовый портрет: {regime}")
    
    # Временные ряды
    plot_time_series(axes[i, 1],
                    param["r1"], param["r2"], 
                    param["K1"], param["K2"],
                    param["alpha12"], param["alpha21"],
                    f"Временные ряды: {regime}")

plt.tight_layout()
plt.show()

# Вывод параметров
print("Параметры модели для варианта 27:")
print("=" * 60)
for regime, param in params.items():
    print(f"\n{regime}:")
    print(f"  r₁ = {param['r1']:.2f}, r₂ = {param['r2']:.2f}")
    print(f"  K₁ = {param['K1']:.2f}, K₂ = {param['K2']:.2f}")
    print(f"  α₁₂ = {param['alpha12']:.2f}, α₂₁ = {param['alpha21']:.2f}")
```
